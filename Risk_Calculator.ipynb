```python
Location_number = {
    "Douala": 38,
    "Yaounde": 57,
    "Buea": 87,
    "Garoua": 40,
    "Limbe": 93,
    "Maroua": 44,
    "Bafoussam": 64,
}
#Define a function to evaluate the risk for a location
def get_safety_status(location_name: str) -> str:
    clean_location = location_name.strip()
    
    if clean_location in Location_number:
        Score = Location_number[clean_location]
        
        if Score < 50:
            return(f"{clean_location}: -> High Risk")
        else:
            return(f"{clean_location}: -> Safe")
    else:
        return(f"{clean_location}: Location not found")
        
#Main function to control program flow
def main():
   #create a list that stores all location entered by user
    User_location_list = []
   
    print("phase1: Data Collection")
    #loop to collect 5 locations
    for i in range(5):
        location = input(f"Enter the location {i + 1} of your choice:")
    #add avery location entered by user inside the list
        User_location_list.append(location)
    
    print("phase2: Processing Risk Status")
#test every location enetered by user
    for location in User_location_list:
        #call the fuction and print the result it return
        Status_result = get_safety_status(location)
        print(Status_result)
    #triggers the main function when you run the cell
#if __name__ == "__main__":
main()
```
