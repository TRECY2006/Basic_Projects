```python
Location_number = {
    "Douala": 38,
    "Yaounde": 57,
    "Buea": 87,
    "Garoua": 40,
    "Limbe": 93,
    "Maroua": 44,
    "Bafoussam": 64,
}

#User_location=input("Enter a locatiion: ")
User_location_list = input("Enter five locations: ").split()
for location in User_location_list:
    clean_location = location.strip()
    
    if clean_location in Location_number:
        Score = Location_number[clean_location]
        if Score < 50:
            print(f"{clean_location}: {Score} points -> High Risk")
        else:
            print(f"{clean_location}: {Score} points -> Safe")
    else:
        print(f"{clean_location}: Location not found")
```
