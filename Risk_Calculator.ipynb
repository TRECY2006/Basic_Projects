```python
Location_number = {
    "Douala": 38,
    "Yaounde": 57,
    "Buea": 87,
    "Garoua": 40,
    "Limbe": 93,
    "Maroua": 44,
    "Bafoussam": 64,
}
#create a list that stores all location entered by user
User_location_list = []
for i in range(5):
    location = input(f"Enter a location {i + 1} of your choice:")
#add avery location entered by user inside the list
    User_location_list.append(location)
#test every location enetered by user
for location in User_location_list:
#clean up any accidental spaces around the location name
    clean_location = location.strip()
    
#check the safety status of the cleaned location name 
    if clean_location in Location_number:
        Score = Location_number[clean_location]
        
        if Score < 50:
            print(f"{clean_location}: -> High Risk")
        else:
            print(f"{clean_location}: -> Safe")
    else:
        print(f"{clean_location}: Location not found")
```
